```python
df_returns=df_returns.rename(columns={'Order ID':'Order_ID','Product ID':'Product_ID'})
```

```python
df_product=df_product.rename(columns={'Product ID':'Product_ID'})
```

```python
df_customer=df_customer.rename(columns={'Customer ID':'Customer_ID'})
```

```python
df_sales=df_sales.rename(columns={'Product ID':'Product_ID', 'Customer ID':'Customer_ID','Order ID':'Order_ID'})
```

```python
df_product['Product_ID'] = range(1,len(df_product)+1)
df_product = df_product.drop('Unnamed: 0',axis=1)
df_product.head()
```

```python
df_sales['Customer_ID'] = range(1,len(df_sales)+1)
df_sales['Product_ID'] = range(1,len(df_sales)+1)
df_sales['Order_ID'] = range(1, len(df_sales)+1)
df_sales = df_sales.drop('Unnamed: 0',axis=1)
df_sales.head()
```

```python
df_returns['Order_ID'] = range(1,len(df_returns)+1)
df_returns['Product_ID'] = range(1,len(df_returns)+1)
df_returns = df_returns.drop('Unnamed: 0',axis=1)
df_returns.head()
```

Creating dimensions

```python
#Product dimension
product_dim = df_product
product_dim.head()
```

```python
#Localisation dimension
localisation_dim = df_customer.iloc[:,-5:]
localisation_dim['Local_ID'] = range(1,len(localisation_dim)+1)
```

```python
localisation_dim.head()
```

```python
date_dim = df_sales.iloc[:,:2]
date_dim = date_dim.rename(columns={'Order Date':'Order_date'})
date_dim = date_dim.drop('Order_ID',axis=1)
date_dim.head()
```

```python
# transform the date dimension into day ,year , month , quarter
date_dim['Order_date'] = pd.to_datetime(date_dim['Order_date'])
date_dim['Year'] = date_dim['Order_date'].dt.year
date_dim['Month'] = date_dim['Order_date'].dt.month
date_dim['Day'] = date_dim['Order_date'].dt.day
date_dim['Quarter'] = date_dim['Order_date'].dt.quarter
```

```python
date_dim['Date_ID'] = range(1,len(date_dim)+1)
```

```python
date_dim.head()
```

```python
date_dim.info()
```

```python
# Cusotmer dimension
customer_dim = df_customer.iloc[:,:3]
customer_dim = customer_dim.rename(columns={'Customer Name' : 'Name'})
customer_dim.head()
```

```python
df_sales['Date_ID'] = range(1,len(date_dim)+1)
df_sales['Local_ID'] = range(1,len(df_sales.drop_duplicates())+1)
df_sales.head()
```

```python
#fact table
fact_sales = df_sales.merge(date_dim, left_on='Date_ID', right_on='Date_ID') \
                       .merge(product_dim, left_on='Product_ID', right_on='Product_ID') \
                       .merge(customer_dim, left_on='Customer_ID', right_on='Customer_ID') \
                       .merge(localisation_dim,left_on='Local_ID',right_on='Local_ID')
fact_sales.head()
```

```python
fact_sales.columns
```

```python
fact_sales = fact_sales[['Order_ID','Product_ID','Local_ID','Date_ID','Customer_ID','Sales','Discount','Quantity','Profit']]
```

```python
localisation_dim.to_csv('local_dim.csv',index=False)
customer_dim.to_csv('customer_dim.csv',index=False)
product_dim.to_csv('product_dim.csv',index=False)
date_dim.to_csv('date_dim.csv',index=False)
fact_sales.to_csv('facts_table.csv',index=False)
```

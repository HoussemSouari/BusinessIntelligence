```python
import pandas as pd
```

Loading the Excel File

```python
df = pd.read_excel('Retail-Supply-Chain-Sales-Dataset.xlsx')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

Splitting the data into source tables

```python
#Sales data source with handling duplicates
sales_table = df[['Order ID', 'Order Date', 'Ship Date', 'Ship Mode', 'Product ID', 
                  'Customer ID', 'Sales', 'Quantity', 'Discount', 'Profit']].drop_duplicates()
sales_table.head()
```

```python
sales_table.to_csv('sales.csv')
```

```python
#Product source with handling duplicates
product_table = df[['Product ID', 'Category', 'Sub-Category', 'Product Name']].drop_duplicates()
product_table.head()
```

```python
product_table.to_csv('Product.csv')
```

```python
#Customer source with handling duplicates
customer_table = df[['Customer ID', 'Customer Name', 'Segment', 'Region', 
                     'Country', 'City', 'State', 'Postal Code']].drop_duplicates()
customer_table.head()
```

```python
customer_table.to_csv('Customer.csv')
```

```python
#Returns source with handling duplicates
returns_table = df[['Order ID', 'Product ID', 'Returned']].drop_duplicates()
returns_table.head()
```

```python
returns_table.to_csv('Returns.csv')
```
